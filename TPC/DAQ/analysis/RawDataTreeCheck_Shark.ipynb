# Endcap display

```python
hRaw = ROOT.TH2F("hRaw",";x [cm];y [cm]", 160,-80,80,160,-80,80)
hNorth = hRaw.Clone("hNorth")
hNorth.GetXaxis().SetTitle(f"Run {run}, North TPC | x [cm]")
hSouth = hRaw.Clone("hSouth")
hSouth.GetXaxis().SetTitle(f"Run {run}, South TPC | x [cm]")
hNorthCount = hRaw.Clone("hNorthCount")
hSouthCount = hRaw.Clone("hSouthCount")
```

```python
SectorPhiMapNorth={}
SectorPhiMapSouth={}

for sector, file in files.items():
    

#     pad.SetLogz()
#     pad.SetLogy()
    print(f"Processing {sector}")
    try:

        n = files[sector].SampleTree.Draw(f"yPos:xPos>>hRaw","(Max$(adcSamples[])  - (adcSamples[0]+adcSamples[1]+adcSamples[2])/3 >50) * (Max$(adcSamples[]) - (adcSamples[0]+adcSamples[1]+adcSamples[2])/3)");
        
        if int(sector)<12:
            hNorth.Add(hRaw, 1)
            SectorPhiMapNorth[sector] = math.atan2(hRaw.GetMean(2), hRaw.GetMean(1))
        else:
            hSouth.Add(hRaw, 1)
            SectorPhiMapSouth[sector] = math.atan2(hRaw.GetMean(2), hRaw.GetMean(1))
            

        n = files[sector].SampleTree.Draw(f"yPos:xPos>>hRaw","(Max$(adcSamples[])  - (adcSamples[0]+adcSamples[1]+adcSamples[2])/3 >50)");
        
        if int(sector)<12:
            hNorthCount.Add(hRaw, 1)
        else:
            hSouthCount.Add(hRaw, 1)
            
        
        print(f"Entries = {n}, hNorth.GetSum()={hNorth.GetSum()}, hSouth.GetSum()={hSouth.GetSum()}, hRaw.GetSum()={hRaw.GetSum()}")
        
    except  Exception as e: 
        print(e) 
        pass
    
hNorth.Divide(hNorthCount)    
hSouth.Divide(hSouthCount)
```

## Log-z scale

```python
RSectorText=25.
SectorText = ROOT.TText(0,0,"")
SectorText.SetTextAlign(22)
SectorText.SetTextSize(15);
```

```python
    
    

c1 = ROOT.TCanvas( 'SectorXY', 'SectorXY', 1200, 600)

c1.Divide(2,1)

pad = c1.cd(1)
pad.SetLogz()
hNorth.DrawClone("colz")

for sector, phi in SectorPhiMapNorth.items():
    SectorText.DrawText(RSectorText * math.cos(phi), RSectorText * math.sin(phi), f"S{sector}")


pad = c1.cd(2)
pad.SetLogz()
hSouth.DrawClone("colz")

for sector, phi in SectorPhiMapSouth.items():
    SectorText.DrawText(RSectorText * math.cos(phi), RSectorText * math.sin(phi), f"S{sector}")
    
c1.Draw()
```

## Linear-z scale

```python
c1 = ROOT.TCanvas( 'SectorXY', 'SectorXY', 1200, 600)

c1.Divide(2,1)

c1.cd(1)
hNorth.SetMaximum(1024)
hNorth.GetZaxis().SetRangeUser(0, 1024);

hNorth.DrawClone("colz").SetMaximum(1024)
for sector, phi in SectorPhiMapNorth.items():
    SectorText.DrawText(RSectorText * math.cos(phi), RSectorText * math.sin(phi), f"S{sector}")

c1.cd(2)
hSouth.SetMaximum(1024)
hSouth.GetZaxis().SetRangeUser(0, 1024);
hSouth.DrawClone("colz").SetMaximum(1024)
for sector, phi in SectorPhiMapSouth.items():
    SectorText.DrawText(RSectorText * math.cos(phi), RSectorText * math.sin(phi), f"S{sector}")
    

c1.Draw()
```
